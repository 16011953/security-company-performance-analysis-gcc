```python
# Import Necessary Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Sample Quantitative Data
quantitative_data = {
    'Respondent': [1, 2, 3, 4, 5],
    'Role': ['Security Company Representative', 'Client (Government Agency)', 'Client (Private Entity)', 'Industry Expert / Stakeholder', 'Client (Individual)'],
    'Country': ['UAE (Dubai)', 'Kuwait', 'Qatar', 'UAE (Dubai)', 'Qatar'],
    'Type of Security Company': ['Private Security', 'Private Security', 'Cybersecurity', 'Consultancy', 'Surveillance'],
    'Overall Satisfaction': ['Satisfied', 'Neutral', 'Very Satisfied', 'Neutral', 'Unsatisfied'],
    'Main Challenges': ['Market Competition, Compliance Costs, Technology Integration', 
                        'Market Competition, Legal Barriers, Operational Issues', 
                        'Technology Integration, Client Retention, Market Competition', 
                        'Operational Issues, Market Competition, Technology Integration', 
                        'Client Retention, Market Competition, Technology Integration'],
}

df = pd.DataFrame(quantitative_data)

# Objective 1: Calculate Percentage Distribution of Security Company Types
type_counts = df['Type of Security Company'].value_counts()
type_percentages = (type_counts / len(df)) * 100

print("Objective 1: Percentage Distribution of Security Company Types")
print(type_percentages)

# Visualization of Security Company Types
plt.figure(figsize=(8, 5))
type_percentages.plot(kind='bar', color='skyblue')
plt.title('Distribution of Security Companies by Type')
plt.xlabel('Type of Security Company')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.show()

# Objective 2: Calculate Percentage of Client Satisfaction Levels
satisfaction_counts = df['Overall Satisfaction'].value_counts()
satisfaction_percentages = (satisfaction_counts / len(df)) * 100

print("\nObjective 2: Percentage of Client Satisfaction Levels")
print(satisfaction_percentages)

# Visualization of Client Satisfaction Levels
plt.figure(figsize=(8, 5))
satisfaction_percentages.plot(kind='bar', color='salmon')
plt.title('Client Satisfaction Levels')
plt.xlabel('Satisfaction Level')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.show()

# Objective 3: Calculate Frequency of Each Challenge
from collections import Counter

# Combine all challenges into a single list
all_challenges = ', '.join(df['Main Challenges']).split(', ')
challenge_counts = Counter(all_challenges)
total_challenges = sum(challenge_counts.values())
challenge_percentages = {k: (v / total_challenges) * 100 for k, v in challenge_counts.items()}

print("\nObjective 3: Frequency and Percentage of Each Challenge")
print(challenge_counts)
print("\nObjective 3: Percentage of Each Challenge")
print(challenge_percentages)

# Visualization of Challenges Faced
plt.figure(figsize=(10, 6))
plt.bar(challenge_percentages.keys(), challenge_percentages.values(), color='lightgreen')
plt.title('Challenges Faced by Security Companies')
plt.xlabel('Challenge')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.show()


```
